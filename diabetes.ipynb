```python
Features = ["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]
X = Diabetic_data[Features]
X.head()
```

```python
output = ["Outcome"]
y = Diabetic_data[output]
y.head()
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
#decision Tree
from sklearn.tree import DecisionTreeRegressor
iowa_model = DecisionTreeRegressor(random_state=1)
iowa_model.fit(X_train, y_train)
Predict_1 = iowa_model.predict(X_val)
print(Predict_1)
print()
print(accuracy_score(Predict_1, y_val))
```

```python
#random Forests
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=70, random_state=0)
clf = clf.fit(X_train, y_train)
Predict_2 = clf.predict(X_val)
print(Predict_2)
print()
print(accuracy_score(Predict_2, y_val))
```

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

x = scale.fit_transform(X_train)

# Add a constant column to our model so we can have a Y-intercept
x = sm.add_constant(X_train)

print (X_train)

est = sm.OLS(y_train, X_train).fit()

print(est.summary())
scaled = scale.transform(X_val)
print(scaled)
predict_3= est.predict(scaled)
print(predict_3)
print()
errors = mean_absolute_error(y_val, predict_3)
print(errors)
```

```python

from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_val_std = sc.transform(X_val)
```

```python
#Support Vector Classifier (SVC)
svc = SVC(C=1.0, random_state=1, kernel='linear')
 
# Fit the model
svc.fit(X_train_std, y_train)
y_predict = svc.predict(X_val_std)
print(y_predict)
print()
print(accuracy_score(y_predict, y_val))
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_estimators = 6)
model.fit(X_train, y_train)
y_pred = model.predict(X_val)
print(y_pred)
print()
print(accuracy_score(y_pred, y_val))
```
